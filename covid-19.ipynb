<h2>Importar dataset diretamente do Github</h2>

```
from pyspark import SparkFiles

url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
sc.addFile(url)
 
path  = SparkFiles.get('owid-covid-data.csv')
df = spark.read.csv("file://" + path, header = True, inferSchema = True, sep = ',').toPandas()
```

```
#Remover valores que não pertencem a um único país e visualizar o ranking, com excessão de World
df_ranking = df.loc[df.location.isin(['High income', 'Asia', 'Upper middle income', 'Europe', 'Africa', 'North America', 'Lower middle income', 'South America', 'European Union','Oceania', 'Low income']) == False] 
deaths_vAbsolute = df_ranking.groupby('location')['total_deaths'].max().sort_values(ascending = False).head(11).map('{:0,.0f}'.format)
deaths_vAbsolute = pd.DataFrame(deaths_vAbsolute).replace(".", ",")
deaths_vAbsolute
```

<h3>Ranking com os 10 países que apresentaram mais mortes por milhão por covid-19</h3>

```
deaths_per_million = df_ranking.groupby('location')['total_deaths_per_million'].max().sort_values(ascending = False).head(10).map('{:0,.0f}'.format)
deaths_per_million = pd.DataFrame(deaths_per_million).replace(".", ",")
deaths_per_million
```

<h3>Ranking com os 10 países que tiveram os maiores percentuais do total de mortes por covid no mundo</h3>

```
#Selecionar valores
world = df_ranking[['location', 'total_deaths']].loc[df_ranking.location == 'World'].max()

countries = df_ranking[['location', 'total_deaths']].loc[df_ranking.location != 'World'].groupby('location').max()
```

```
#Visualizar o ranking
import pandas as pd

percent_deaths = (countries.total_deaths/world.total_deaths).sort_values(ascending = False).head(10)
rank_population = pd.DataFrame(rank_population).replace(".", ",")
percent_deaths = pd.DataFrame(percent_deaths).style.format({'total_deaths': "{:.2%}"})
percent_deaths
```

<h3>Ranking de países mais populosos</h3>

```
rank_population = df_ranking.groupby('location')['population'].max().sort_values(ascending = False).head(10).map('{:0,.0f}'.format)
rank_population = pd.DataFrame(rank_population).replace(".", ",")
rank_population
```
